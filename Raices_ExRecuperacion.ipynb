```python
# Encuentre la o las raíces de la ecuación planteada

import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import fsolve

# Define the function
def f(x):
  return x**3 + 2*x**2 - 5

# Generate x values
x = np.linspace(-2, 2, 400)  # Adjust the range as needed

# Calculate corresponding y values
y = f(x)

# Find the root using fsolve
root = fsolve(f, 1) # 1 is an initial guess

# Create the plot
plt.plot(x, y)

# Mark the root on the plot
plt.plot(root, f(root), 'ro', markersize=8)  # 'ro' for red circle marker

# Customize the plot (optional)
plt.title("Plot of f(x) = x^3 + 2x^2 - 5")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)

# Add text indicating the root
plt.text(root + 0.1, f(root), f"Root: {root[0]:.4f}", fontsize=10)

# Display the plot
plt.show()

print(f"The root of the function is approximately: {root[0]}")
```

```python
# prompt: Compare la solución para dos métodos diferentes vistos en clase.

import matplotlib.pyplot as plt
import numpy as np

# Define the function
def f(x):
  return x**3 + 2*x**2 - 5

# Bisection method
def bisection(a, b, tolerance=1e-6, max_iterations=100):
    if f(a) * f(b) >= 0:
        print("Bisection method fails.")
        return None

    iterations = 0
    while (b - a) / 2 > tolerance and iterations < max_iterations:
        c = (a + b) / 2
        if f(c) == 0:
            return c
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
        iterations += 1
    return (a + b) / 2

# Secant method
def secant(x0, x1, tolerance=1e-6, max_iterations=100):
    iterations = 0
    while abs(x1 - x0) > tolerance and iterations < max_iterations:
        x2 = x1 - f(x1) * (x1 - x0) / (f(x1) - f(x0))
        x0 = x1
        x1 = x2
        iterations += 1
    return x1

# Generate x values for plotting
x = np.linspace(-2, 1.5, 400)
y = f(x)

# Find roots using both methods
root_bisection = bisection(1, 2)  # Initial interval [1, 2]
root_secant = secant(1, 2)      # Initial guesses 1 and 2


# Plot the function and the roots
plt.plot(x, y, label='f(x)')
plt.plot(root_bisection, f(root_bisection), 'ro', markersize=8, label='Bisection Root')
plt.plot(root_secant, f(root_secant), 'go', markersize=8, label='Secant Root')


plt.title('Roots of f(x) = x^3 + 2x^2 - 5')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

# Add text annotations
plt.text(root_bisection + 0.1, f(root_bisection), f"Bisection: {root_bisection:.4f}", fontsize=10)
plt.text(root_secant - 1, f(root_secant), f"Secant: {root_secant:.4f}", fontsize=10)

plt.show()

print(f"Root found by bisection method: {root_bisection}")
print(f"Root found by secant method: {root_secant}")
```

```python
#Determine el número de iteraciones de ambos métodos hasta la convergencia.

# Bisection method
def bisection_it(a, b, tolerance=1e-6, max_iterations=100):
    if f(a) * f(b) >= 0:
        print("Bisection method fails.")
        return None, 0 #return None and the number of iterations

    iterations = 0
    while (b - a) / 2 > tolerance and iterations < max_iterations: #Criterio de parada
        c = (a + b) / 2
        if f(c) == 0:
            return c, iterations
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
        iterations += 1
    return (a + b) / 2, iterations

# Secant method
def secant_it(x0, x1, tolerance=1e-6, max_iterations=100):
    iterations = 0
    while abs(x1 - x0) > tolerance and iterations < max_iterations: #Criterio de parada
        x2 = x1 - f(x1) * (x1 - x0) / (f(x1) - f(x0))
        x0 = x1
        x1 = x2
        iterations += 1
    return x1, iterations

# Example usage
root_bisection, iterations_bisection = bisection_it(1, 2)
root_secant, iterations_secant = secant_it(1,2)

print(f"Root found by bisection method: {root_bisection} in {iterations_bisection} iterations")
print(f"Root found by secant method: {root_secant} in {iterations_secant} iterations")
```
