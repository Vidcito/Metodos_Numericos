```python
import matplotlib.pyplot as plt

def graficar_trayectorias(trayectoria_jacobi, trayectoria_seidel):
    x_jacobi = [float(punto[0]) for punto in trayectoria_jacobi]
    y_jacobi = [float(punto[1]) for punto in trayectoria_jacobi]

    x_seidel = [float(punto[0]) for punto in trayectoria_seidel]
    y_seidel = [float(punto[1]) for punto in trayectoria_seidel]

    plt.figure(figsize=(8, 6))
    plt.plot(x_jacobi, y_jacobi, marker="s", linestyle="-", label="Método Jacobi", color="blue")
    plt.plot(x_seidel, y_seidel, marker="^", linestyle="--", label="Método Seidel", color="green")
    plt.title("Comparación de trayectorias")
    plt.xlabel("$x_1$")
    plt.ylabel("$x_2$")
    plt.legend()
    plt.grid(alpha=0.5)
    plt.show()

def graficar_trayectoria_individual(trayectoria, metodo="Método"):
    x_coords = [float(punto[0]) for punto in trayectoria]
    y_coords = [float(punto[1]) for punto in trayectoria]

    plt.figure(figsize=(8, 6))
    plt.plot(x_coords, y_coords, marker="o", linestyle="-", label=f"Trayectoria - {metodo}")
    plt.title(f"Trayectoria del {metodo}")
    plt.xlabel("$x_1$")
    plt.ylabel("$x_2$")
    plt.legend()
    plt.grid(alpha=0.5)
    plt.show()

# Parámetros iniciales
A = [[2, 10], [3, 2]]
b = [16, 11]
x0 = [1, 1]

# Ejecutar métodos
sol_jacobi, trayectoria_jacobi = gauss_jacobi(A=A, b=b, x0=x0, tol=1e-4, max_iter=200)
sol_seidel, trayectoria_seidel = gauss_seidel(A=A, b=b, x0=x0, tol=1e-4, max_iter=200)

# Graficar trayectorias
graficar_trayectorias(trayectoria_jacobi, trayectoria_seidel)

# Graficar trayectorias individuales
graficar_trayectoria_individual(trayectoria_jacobi, metodo="Gauss-Jacobi")
graficar_trayectoria_individual(trayectoria_seidel, metodo="Gauss-Seidel")
```

```python
# Converge con x0 = [1, 1]
x, tray = gauss_jacobi(A=[[3, 2], [2, 10]], b=[11, 16], x0=[1, 1], tol=0.0001, max_iter=200)
x2, tray2 = gauss_seidel(A=[[3, 2], [2, 10]], b=[11, 16], x0=[1, 1], tol=0.0001, max_iter=200)
graficar_trayectorias(tray, tray2)
```
