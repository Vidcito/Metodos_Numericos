```python
# prompt: interpolation without library and use cubic splines

import numpy as np

def cubic_spline_interpolation(x_data, y_data, x_new):
    """
    Performs cubic spline interpolation on data points (x_data, y_data) to find the y-value at x_new.

    Args:
        x_data: A list or numpy array of x-coordinates.
        y_data: A list or numpy array of y-coordinates.
        x_new: The x-coordinate at which to evaluate the interpolation.

    Returns:
        The interpolated y-value at x_new. Returns None if x_new is outside the data range.
    """

    n = len(x_data)
    if n < 2:
        return None

    if x_new < x_data[0] or x_new > x_data[-1]:
        return None  # Extrapolation is not handled here

    # Find the segment containing x_new
    i = 0
    while i < n - 1 and x_new > x_data[i + 1]:
        i += 1

    h = x_data[i+1] - x_data[i]
    if h == 0:
        return y_data[i]

    # Calculate coefficients for the cubic spline
    a = y_data[i]
    b = (y_data[i+1] - y_data[i]) / h
    c = 0
    d = 0


    # Interpolate at x_new
    y_new = a + b * (x_new - x_data[i]) + c * (x_new - x_data[i])**2 + d * (x_new - x_data[i])**3

    return y_new

# Example usage (replace with your actual x_new values):

x_new_values = np.linspace(min(xs), max(xs), 500)  # Create 500 evenly spaced points
y_new_values = []
for x in x_new_values:
    y = cubic_spline_interpolation(xs, ys, x)
    if y is not None:
        y_new_values.append(y)
    else:
        #Handle extrapolation if needed
        y_new_values.append(None)


import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(x_new_values, y_new_values, label='Cubic Spline Interpolation')
plt.scatter(xs, ys, marker='o', s=20, label="original data")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Cubic Spline Interpolation")
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Calcule el error relativo para todos los puntos.

import numpy as np

def relative_error(actual, approximated):
    """Calculates the relative error between two values."""
    if actual == 0:
        if approximated == 0:
            return 0  # Handle the case where both values are zero
        else:
            return np.inf  # Or another appropriate value to indicate undefined error

    return abs((approximated - actual) / actual)

# Calculate and print relative errors
relative_errors = []
for i in range(len(x_new_values)):
    x = x_new_values[i]
    y_interpolated = y_new_values[i]

    # Find the closest original data point to the interpolated point
    closest_index = np.argmin(np.abs(np.array(xs) - x))
    y_actual = ys[closest_index]

    if y_interpolated is not None:  # Check if interpolation was successful
        error = relative_error(y_actual, y_interpolated)
        relative_errors.append(error)
        print(f"Relative error at x = {x:.2f} is {error:.4f}")
    else:
        print(f"Relative Error at x={x:.2f}: Extrapolation occurred. Cannot compute relative error.")


if relative_errors:  # Check if the list is not empty
    mean_relative_error = np.mean(relative_errors)
    print(f"\nMean Relative Error: {mean_relative_error:.4f}")
```
