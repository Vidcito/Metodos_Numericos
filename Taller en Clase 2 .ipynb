```python
# prompt: Codigo de la funcion coseno y 5 aproximaciones de taylor con grafico las de taylos lineas punteadas

import numpy as np
import matplotlib.pyplot as plt

def cos_taylor(x, n_terms):
  approx = 0
  for i in range(n_terms):
    approx += ((-1)**i) * (x**(2*i)) / np.math.factorial(2*i)
  return approx

x = np.linspace(-4*np.pi, 4*np.pi, 250)

y_cos = np.cos(x)

y_taylor1 = [cos_taylor(xi, 1) for xi in x]
y_taylor2 = [cos_taylor(xi, 2) for xi in x]
y_taylor3 = [cos_taylor(xi, 3) for xi in x]
y_taylor4 = [cos_taylor(xi, 4) for xi in x]
y_taylor5 = [cos_taylor(xi, 5) for xi in x]


plt.figure(figsize=(10, 6))
plt.plot(x, y_cos, label='cos(x)')
plt.plot(x, y_taylor1, '--', label='Aproximación Taylor (1 término)')
plt.plot(x, y_taylor2, '--', label='Aproximación Taylor (2 términos)')
plt.plot(x, y_taylor3, '--', label='Aproximación Taylor (3 términos)')
plt.plot(x, y_taylor4, '--', label='Aproximación Taylor (4 términos)')
plt.plot(x, y_taylor5, '--', label='Aproximación Taylor (5 términos)')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Aproximación de cos(x) usando la serie de Taylor')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# prompt: Codigo de la funcion 1/(1-x) y 5 aproximaciones de taylor con grafico las de taylos lineas punteadas

import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return 1 / (1 - x)

def taylor_approx(x, n_terms):
  approx = 0
  for i in range(n_terms):
    approx += x**i
  return approx

x = np.linspace(-0.9, 0.9, 250)

y_f = f(x)

y_taylor1 = [taylor_approx(xi, 1) for xi in x]
y_taylor2 = [taylor_approx(xi, 2) for xi in x]
y_taylor3 = [taylor_approx(xi, 3) for xi in x]
y_taylor4 = [taylor_approx(xi, 4) for xi in x]
y_taylor5 = [taylor_approx(xi, 5) for xi in x]

plt.figure(figsize=(10, 6))
plt.plot(x, y_f, label='1/(1-x)')
plt.plot(x, y_taylor1, '--', label='Aproximación Taylor (1 término)')
plt.plot(x, y_taylor2, '--', label='Aproximación Taylor (2 términos)')
plt.plot(x, y_taylor3, '--', label='Aproximación Taylor (3 términos)')
plt.plot(x, y_taylor4, '--', label='Aproximación Taylor (4 términos)')
plt.plot(x, y_taylor5, '--', label='Aproximación Taylor (5 términos)')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Aproximación de 1/(1-x) usando la serie de Taylor')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def ln_taylor(x, n_terms):
  approx = 0
  for i in range(1, n_terms + 1):
    approx += ((-1)**(i+1)) * ((x-1)**i) / i
  return approx

x = np.linspace(0, 2, 250)

y_ln = np.log(x)

y_taylor1 = [ln_taylor(xi, 1) for xi in x]
y_taylor2 = [ln_taylor(xi, 2) for xi in x]
y_taylor3 = [ln_taylor(xi, 3) for xi in x]
y_taylor4 = [ln_taylor(xi, 4) for xi in x]
y_taylor5 = [ln_taylor(xi, 5) for xi in x]


plt.figure(figsize=(10, 6))
plt.plot(x, y_ln, label='ln(x)')
plt.plot(x, y_taylor1, '--', label='Aproximación Taylor (1 término)')
plt.plot(x, y_taylor2, '--', label='Aproximación Taylor (2 términos)')
plt.plot(x, y_taylor3, '--', label='Aproximación Taylor (3 términos)')
plt.plot(x, y_taylor4, '--', label='Aproximación Taylor (4 términos)')
plt.plot(x, y_taylor5, '--', label='Aproximación Taylor (5 términos)')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Aproximación de ln(x) usando la serie de Taylor')
plt.legend()
plt.grid(True)
plt.show()
```
